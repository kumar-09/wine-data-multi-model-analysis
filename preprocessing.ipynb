# Data Preprocessing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

plt.style.use('seaborn-v0_8-whitegrid')
sns.set(font_scale=1.2)

# 1. Load datasets
wine_red = pd.read_excel("winequality_red.xlsx")
wine_white = pd.read_excel("winequality_white.xlsx")

print("Red wine dataset shape:", wine_red.shape)
print("White wine dataset shape:", wine_white.shape)

# Function to perform data analysis on wine datasets
def analyze_wine_data(df, wine_type_name):
    print(f"\n{'='*50}")
    print(f"ANALYZING {wine_type_name.upper()} WINE")
    print(f"{'='*50}")
    
    # 1. Basic exploration
    print(f"\n{wine_type_name} wine dataset shape:", df.shape)
    print(f"\nFirst 5 rows of {wine_type_name} wine data:")
    display(df.head())
    
    print(f"\nBasic statistics for {wine_type_name} wine:")
    display(df.describe())
    
    print(f"\nMissing values in {wine_type_name} wine data:")
    display(df.isnull().sum())
    
    # 2. Quality distribution visualization
    print(f"\nClass distribution (Quality) for {wine_type_name} wine:")
    quality_counts = df['quality'].value_counts().sort_index()
    display(quality_counts)
    print("Percentage distribution:")
    display(round(quality_counts / len(df) * 100, 2))
    
    # Visualize class distribution
    plt.figure(figsize=(10, 5))
    ax = sns.countplot(x='quality', data=df)
    plt.title(f'{wine_type_name} Wine Quality Distribution')
    plt.xlabel('Quality Score')
    plt.ylabel('Count')
    for container in ax.containers:
        ax.bar_label(container)
    plt.tight_layout()
    plt.show()
    
    # 3. Correlation analysis heatmap
    plt.figure(figsize=(12, 10))
    corr = df.corr()
    mask = np.triu(np.ones_like(corr, dtype=bool))
    sns.heatmap(corr, annot=True, mask=mask, cmap='RdBu_r', vmin=-1, vmax=1, fmt='.2f')
    plt.title(f'{wine_type_name} Wine Feature Correlation Matrix')
    plt.tight_layout()
    plt.show()
    
    # 4. Box plots with Z-score outlier counts
    numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
    numeric_cols.remove('quality')  
    
    plt.figure(figsize=(15, 10))
    z_outlier_counts = {}
    for i, col in enumerate(numeric_cols, 1):
        plt.subplot(3, 4, i)
        z_scores = np.abs(stats.zscore(df[col]))
        outliers = (z_scores > 3).sum()
        z_outlier_counts[col] = outliers
        sns.boxplot(y=col, data=df)
        plt.title(f'{col} (Z-score outliers: {outliers})')
    plt.tight_layout()
    plt.show()
    
    print(f"\n{wine_type_name} Wine Z-Score Outlier Summary:")
    for col, count in z_outlier_counts.items():
        print(f"{col}: {count} outliers ({count/len(df):.2%} of data)")
    
    # 5. Feature correlation with quality target
    target_corr = df[numeric_cols].corrwith(df['quality']).sort_values(ascending=False)
    print(f"\n{wine_type_name} Wine feature correlation with quality target:")
    display(target_corr)
    
    # Top 5 most correlated features visualization
    plt.figure(figsize=(12, 8))
    top_features = target_corr.abs().sort_values(ascending=False).head(5).index
    for i, feature in enumerate(top_features, 1):
        plt.subplot(2, 3, i)
        sns.regplot(x=feature, y='quality', data=df)
        plt.title(f'{feature} vs Quality (corr: {target_corr[feature]:.3f})')
    plt.tight_layout()
    plt.show()
    
    # 6. Feature box plots by quality levels
    plt.figure(figsize=(20, 15))
    for i, feature in enumerate(numeric_cols, 1):
        plt.subplot(4, 3, i)
        sns.boxplot(x='quality', y=feature, data=df)
        plt.title(f'{feature} by Quality Rating')
        plt.xlabel('Quality Score')
        plt.ylabel(feature)
        
        y_min = df[feature].min() * 0.9
        y_max = df[feature].max() * 1.1
        plt.ylim(y_min, y_max)
    
    plt.tight_layout()
    plt.suptitle(f'{wine_type_name} Wine: Feature Distributions Across Quality Levels', fontsize=16)
    plt.subplots_adjust(top=0.95)
    plt.show()

    return df

# Processing each dataset separately
red_wine_analyzed = analyze_wine_data(wine_red, "Red")
white_wine_analyzed = analyze_wine_data(wine_white, "White")

# Key characteristics between red and white wines
print("\n" + "="*50)
print("COMPARING RED AND WHITE WINE CHARACTERISTICS")
print("="*50)

# Median values for red and white wines
red_wine_medians = wine_red.median()
white_wine_medians = wine_white.median()
print("Median values for red wine features:")
print(red_wine_medians)
print("\nMedian values for white wine features:")
print(white_wine_medians)

# Plot distributions of most important features
combined_impacts = {}
for feature in wine_red.columns:
    if feature not in ['quality', 'wine_type']:
        red_impact = abs(wine_red.corr()['quality'][feature])
        white_impact = abs(wine_white.corr()['quality'][feature])
        combined_impacts[feature] = max(red_impact, white_impact)

# Top 4 features
top_features = sorted(combined_impacts.items(), key=lambda x: x[1], reverse=True)[:4]
top_feature_names = [feature for feature, impact in top_features]

# Plot comparisons
plt.figure(figsize=(15, 12))
for i, feature in enumerate(top_feature_names, 1):
    plt.subplot(2, 2, i)
    
    # Plot the PDFs for red and white wines
    sns.kdeplot(wine_red[feature], label='Red wine', color='darkred')
    sns.kdeplot(wine_white[feature], label='White wine', color='darkgreen')
    

    r_corr = round(wine_red.corr()['quality'][feature], 3)
    w_corr = round(wine_white.corr()['quality'][feature], 3)
    
    plt.title(f'{feature} (Red corr: {r_corr}, White corr: {w_corr})')
    plt.xlabel(feature)
    plt.ylabel('Probability Density')
    plt.legend()

plt.suptitle('Distribution of Top 4 Most Impactful Features: Red vs White Wines', fontsize=16)
plt.tight_layout()
plt.subplots_adjust(top=0.92)
plt.show()

# Comparing common characteristics
common_cols = ['alcohol', 'volatile acidity', 'density', 'total sulfur dioxide']
plt.figure(figsize=(14, 10))
for i, col in enumerate(common_cols, 1):
    plt.subplot(2, 2, i)
    sns.kdeplot(wine_red[col], label='Red wine')
    sns.kdeplot(wine_white[col], label='White wine')
    plt.title(f'Distribution of {col}')
    plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Box plots showing feature distributions across quality levels
plt.figure(figsize=(20, 15))

#List of all numeric features excluding quality
features = [col for col in wine_red.columns if col != 'quality']

#Subplot for each feature
for i, feature in enumerate(features, 1):
    plt.subplot(4, 3, i)
    
    # Boxplot
    sns.boxplot(x='quality', y=feature, data=wine_red)
    
    
    plt.title(f'{feature} by Quality Rating')
    plt.xlabel('Quality Score')
    plt.ylabel(feature)
    
    
    y_min = wine_red[feature].min() * 0.9
    y_max = wine_red[feature].max() * 1.1
    plt.ylim(y_min, y_max)

plt.tight_layout()
plt.suptitle('Red Wine: Feature Distributions Across Quality Levels', fontsize=16)
plt.subplots_adjust(top=0.95)
plt.show()

# Repeat for white wine
plt.figure(figsize=(20, 15))

features = [col for col in wine_white.columns if col != 'quality']

for i, feature in enumerate(features, 1):
    plt.subplot(4, 3, i)
    sns.boxplot(x='quality', y=feature, data=wine_white)
    plt.title(f'{feature} by Quality Rating')
    plt.xlabel('Quality Score')
    plt.ylabel(feature)
    
    y_min = wine_white[feature].min() * 0.9
    y_max = wine_white[feature].max() * 1.1
    plt.ylim(y_min, y_max)

plt.tight_layout()
plt.suptitle('White Wine: Feature Distributions Across Quality Levels', fontsize=16)
plt.subplots_adjust(top=0.95)
plt.show()
```
